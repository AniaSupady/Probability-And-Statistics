# Introduction

Below is the topic for the challenge:





"What is the difference between mutually exclusive events and independent events?"


# The solution

Let's start tossing a coin. We know very well that the probablity of getting a Head is 50% or 0.5 or 1/2.

The same goes to getting a Tail. The the probablity of getting a Tail is 50% or 0.5 or 1/2.


If we were asked to draw a ven diagram, they would look like this:

## Ven Diagram (disjoint)

The ven diagrams have no intersection between each event. This means, you can only get Head or Tail. You cannot get both at the same time. We say that events Head and Tail are mutually exclusive or disjoint.

##  Dependent Events?

The occurence of one event depends on the fact that the other event did not occure. In this sense, the two events are dependent.

If you know that event Head accured, then you know automatically that the event Tail did not occure.

Therefore, the knowledge of one event helps you to understand the other event.

## Math

Lets show mathematically that the event Head and Tail are dependent:

>The left side reads: Joint probability of getting Head and Tail.
                      The intersection of Head and Tail is 0,
                      Therefore, the probability of Head and Tail is just 0.


>The right side reads: Marginal probability of getting Head times the marginal probability of getting Tail, which comes to 1/2 * 1/2.


>If the left and right side are not equal, then these two events are **dependent**.  

$$ P(A\cap B) \neq P(A)\cdot P(B) $$
$$0 = P(A\cap B) \neq P(A)\cdot P(B) = \frac{1}{4}$$

$$0 \neq \frac{1}{2}* \frac{1}{2}$$

$$0 \neq \frac{1}{4}$$

## Independent Outcomes?

Although the occurence of these events is dependent in each other, but the probability of the either event occuring is not dependent on occurence proceeding or following event.

In other words, the chances of Get Head do not change based on getting Head once before.

Most logical example to think of is mother giving birth to boy or a girl. The chances of having a girl are not improved by the fact that mother already had a girl.

# Simulation

Why do you believe that the probability of getting a Tail in a fair coin is equal to  1/2 ?


Probability = Long Run Relative Frequency

The Law of Large Numbers (LLN): The relative frequency of an outcome converges to a number, as the number of observed outcomes increases.

LLN assumes that events are independent.

If you toss a coin thousands of times, the proportion of Tails will be likely closer to  12 .

Let's toss a coin 10,000 times and count how many Tails:

```python
# Library to generate random numbers
import random

# Simulation
result= []

for toss in range(10000):
   coin = random.choice(['Head', 'Tail'])
   result.append(coin)

result.count('Tail')
```

```python
# Simulation
n=1
prob=[]
flip=[]

while n < 2000:
    head = 0
    tail = 0

    for i in range(n):
        if random.randint(0,1) == 0:
            head+=1
        else:
            tail+=1

    k = head/(head+tail)
    prob.append(k)
    flip.append(n)
    n+= 1
```

```python
# Plot chart
import plotly.express as px
fig = px.scatter( x = flip, y = prob,
             labels={"x": "Number of Flips",
                     "y": "Probability of Heads"} )

# Horizontal Line
fig.update_layout( shapes = [
    dict(
      type= 'line',
      yref= 'y', y0= 0.5, y1= 0.5,
      xref= 'x', x0= 0, x1= 2000    )])


fig.show()
```
